### I will test the function `grov_itera` defined in the `Grover_algorithm_iterations.py` file with different binary number string by plotting the circuit and run it on the `qasm simulator`:

```python
from qiskit.tools.visualization import plot_histogram
%run Grover_algorithm_iterations_update.py
```

## Case of one solution:

```python
############ String to be found '111' using noancilla type ############
test_1=grover_itera(['111'], 'noancilla', 2)
test_1.decompose().draw('mpl')
```

```python
simulator=Aer.get_backend('qasm_simulator')
count1=execute(test_1,simulator).result().get_counts()
print(count1)
plot_histogram(count1,color='limegreen')
```

## The case of two solutions (I will use the `noancilla` circuit):
